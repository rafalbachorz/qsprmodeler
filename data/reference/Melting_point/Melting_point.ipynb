```python
data = pd.read_excel('data_raw.csv')
```

```python
import pubchempy as pcp
```

```python
compounds_l = []
for name in data['NAME']:
    try:
        
        compound = pcp.get_compounds(name, 'name')
        compounds_l.append(compound)
    except:
        
        print("Not found : "+str(name))
compounds_l[0:5]
```

```python
data['ID'] = compounds_l
```

```python
data.head()
```

```python
isomeric_smiles = []
for compound in data['ID']:
    try:
        
        isomeric_smiles.append(compound[0].isomeric_smiles)
    except:
        isomeric_smiles.append('ERROR')
isomeric_smiles[0:5]
```

```python
isomeric_smiles
```

```python
data['SMILES'] = isomeric_smiles
```
