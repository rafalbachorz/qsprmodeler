```python
import pandas as pd
pd.options.display.max_colwidth = 300
pd.options.display.max_columns = 300
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt

DATA_DIR=Path("../data")
nm2mol = 1E9
```

```python
df_IC50 = pd.read_csv(DATA_DIR/"processed"/"ar_homo_sapiens_chembl_3.3_more_ChEMBL_columnsIC50.csv_processed_expert.csv", sep=",")
df_EC50 = pd.read_csv(DATA_DIR/"processed"/"ar_homo_sapiens_chembl_3.3_more_ChEMBL_columnsEC50.csv_processed_expert.csv", sep=",")
```

```python
df = pd.concat([df_IC50, df_EC50], axis=0)
```

```python
df.columns.tolist()
```

```python
replace_dir = {"ACTIVE": 1, "INACTIVE": 0}
```

```python
df["class_activity_median_expert_1000"] = df["class_activity_median_expert_1000"].replace(replace_dir)
```

```python
df_to_be_exported = df[["coms_canonical_smiles", "act_median_expert", "class_activity_median_expert_1000"]]
```

```python
df_to_be_exported.columns = ["coms_canonical_smiles", "act_median_expert_0", "class_activity_median_expert_1000_0"]
```

```python
df_to_be_exported.to_csv(DATA_DIR/"processed"/"ar_homo_sapiens_chembl_3.3_more_ChEMBL_columnsIC50EC50_combined_expert.csv", sep=",", index=False)
```
