```python
def get_max_sim(query_smiles, reference):
    fp_query = AllChem.GetMorganFingerprint(Chem.MolFromSmiles(query_smiles), 2)
    return max(DataStructs.BulkTanimotoSimilarity(fp_query, fps))
```
